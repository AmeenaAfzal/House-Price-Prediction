```python
import matplotlib.pyplot as plt
plt.plot(x,y,linestyle="--",marker="o")
plt.grid()
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.hist(x,bins=20)
plt.show()
```
